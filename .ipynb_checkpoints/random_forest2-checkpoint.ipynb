```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
#Read in csv 
df = pd.read_csv("data/exoplanet_data.csv")

# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')

# Drop the null rows
df = df.dropna()

df.head(60)
```

# Select your features (columns)

```python
list(df.columns)
```

```python
# Set features. This will also be used as your x values. 
# First trained model with all features, then used feature_importance to reduce to 5 most important features

# selected_features = df.drop("koi_disposition", axis=1)
selected_features = df[["koi_fpflag_co", "koi_fpflag_nt", "koi_fpflag_ss", "koi_model_snr", "koi_prad"]]
selected_features
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
# Assign the data to X and y
X = selected_features

y = df['koi_disposition']

print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=22)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale the data
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical

X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)

print(y_train_categorical)
```

# Train the Model



```python
#Random Forest algorithm
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200, random_state=22)

#Fit the model to the data
rf = rf.fit(X_train_scaled, y_train_categorical)

#Quantify the model
rf.score(X_test_scaled, y_test_categorical)
```

```python
sorted(zip(rf.feature_importances_, selected_features), reverse=True)
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
# Guidance provided by https://www.kaggle.com/sociopath00/random-forest-using-gridsearchcv
from sklearn.model_selection import GridSearchCV
param_grid = { 
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}

CV_rf = GridSearchCV(estimator=rf, param_grid=param_grid, cv= 5)
```

```python
# Train the model with GridSearch
CV_rf.fit(X_train_scaled, y_train_categorical)
```

```python
#Get best params
print(CV_rf.best_params_)
print(CV_rf.best_score_)
```

```python
#Create new RF model with best params
rf2 = RandomForestClassifier(random_state=22, max_features='auto', n_estimators= 500, max_depth=6, criterion='gini')

#Fit the model to the data
rf2 = rf2.fit(X_train_scaled, y_train_categorical)

#Quantify the model
rf2.score(X_test_scaled, y_test_categorical)
```

# Save the Model

```python
# save your model by updating "your_name" with your name
# and "your_model" with your model variable
# be sure to turn this in to BCS
# if joblib fails to import, try running the command to install in terminal/git-bash
import joblib
filename = 'julia_obrien.sav'
joblib.dump(rf2, filename)
```
