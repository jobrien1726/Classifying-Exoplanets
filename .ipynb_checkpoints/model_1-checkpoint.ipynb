# Read the CSV and Perform Basic Data Cleaning

```python
#Read in csv 
df = pd.read_csv("data/exoplanet_data.csv")

# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')

# Drop the null rows
df = df.dropna()

df.head()
```

# Select your features (columns)

```python
list(df.columns)
```

```python
# Set features. This will also be used as your x values.
selected_features = df.drop("koi_disposition", axis=1)
selected_features

# selected_features = df[['koi_fpflag_ss','koi_period','koi_time0bk','koi_impact','koi_duration','koi_depth','koi_prad','koi_teq','koi_insol','koi_model_snr','koi_tce_plnt_num','koi_steff','koi_slogg','koi_srad','ra','dec','koi_kepmag']]
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
# Assign the data to X and y
X = selected_features

y = df['koi_disposition']

print(X.shape, y.shape)
```

```python
#Create train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=22)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale the data
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical

X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)

print(y_train_categorical)
```

# Train the Model



```python
#Random Forest algorithm
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)

#Fit the model to the data
rf = rf.fit(X_train, y_train_categorical)

#Quantify the model
rf.score(X_test, y_test_categorical)
```
